```r
#include <iostream>
using namespace std;

double circleArea(double r) { return 3.14 * r * r; }
double squareArea(double a) { return a * a; }
double rectangleArea(double a, double b) { return a * b; }
double triangleArea(double a, double h) { return 0.5 * a * h; }

int main() {
    int choice;
    double a, b;

    cout << "Калькулятор площ\n";
    cout << "1 - Коло\n";
    cout << "2 - Квадрат\n";
    cout << "3 - Прямокутник\n";
    cout << "4 - Трикутник\n";
    cout << "Ваш вибір: ";
    cin >> choice;

    switch (choice) {
    case 1:
        cout << "Введіть радіус: ";
        cin >> a;
        cout << "Площа кола = " << circleArea(a) << endl;
        break;
    case 2:
        cout << "Введіть сторону: ";
        cin >> a;
        cout << "Площа квадрата = " << squareArea(a) << endl;
        break;
    case 3:
        cout << "Введіть сторони: ";
        cin >> a >> b;
        cout << "Площа прямокутника = " << rectangleArea(a, b) << endl;
        break;
    case 4:
        cout << "Введіть основу і висоту: ";
        cin >> a >> b;
        cout << "Площа трикутника = " << triangleArea(a, b) << endl;
        break;
    default:
        cout << "Невірний вибір!";
    }
    return 0;
}
```
